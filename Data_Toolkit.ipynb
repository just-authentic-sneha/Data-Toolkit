# Data Toolkit




## 1. What is NumPy, and why is it widely used in Python?
NumPy is a numerical computing library. It offers fast, memory-efficient arrays and functions for math, stats, and logic. It's foundational for many data science libraries.

---

## 2. How does broadcasting work in NumPy?
Broadcasting lets NumPy perform operations on arrays of different shapes by expanding smaller arrays to match the shape of larger ones.

---

## 3. What is a Pandas DataFrame?
A 2D tabular data structure with labeled rows and columns, like an Excel sheet or SQL table.

---

## 4. Explain the use of the `groupby()` method in Pandas.
It splits data into groups, applies a function to each group, and combines the results — ideal for aggregation and summarization.

---

## 5. Why is Seaborn preferred for statistical visualizations?
Seaborn creates beautiful and informative statistical plots with less code and has built-in support for complex visualizations like violin or regression plots.

---

## 6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays are faster, use less memory, and support vectorized operations.
- Python lists are flexible but slower and don't support element-wise math.

---

## 7. What is a heatmap, and when should it be used?
A heatmap shows data in a matrix using color to indicate values. Use it for correlations, confusion matrices, or density analysis.

---

## 8. What does the term “vectorized operation” mean in NumPy?
Operations are applied to entire arrays without loops — it’s faster and more readable.

---

## 9. How does Matplotlib differ from Plotly?
- Matplotlib is static, good for print.
- Plotly is interactive, good for dashboards and web apps.

---

## 10. What is the significance of hierarchical indexing in Pandas?
It allows multiple levels of indexing (MultiIndex), which is helpful for grouped or multi-dimensional data in tabular format.

---

## 11. What is the role of Seaborn’s `pairplot()` function?
It shows pairwise relationships using scatter plots and histograms/KDEs — great for quick exploratory data analysis (EDA).

---

## 12. What is the purpose of the `describe()` function in Pandas?
It returns summary statistics (mean, std, min, max, etc.) for numerical columns.

---

## 13. Why is handling missing data important in Pandas?
Missing values can skew results or cause errors. Pandas offers `isnull()`, `dropna()`, `fillna()` to detect and handle them.

---

## 14. What are the benefits of using Plotly for data visualization?
- Interactive charts
- Support for 3D, maps, and dashboards
- Ideal for web-based analytics and reporting

---

## 15. How does NumPy handle multidimensional arrays?
It uses `ndarrays`, which support operations across any number of dimensions with tools like `reshape()`, `transpose()`, etc.

---

## 16. What is the role of Bokeh in data visualization?
Bokeh is for interactive plots and dashboards in the browser using Python — it’s great for web apps.

---

## 17. Explain the difference between `apply()` and `map()` in Pandas.
- `map()` is for element-wise operations on Series.
- `apply()` can be used on DataFrames or Series and supports row/column-wise functions.

---

## 18. What are some advanced features of NumPy?
- Broadcasting
- Linear algebra (`linalg`)
- Random sampling
- Memory mapping
- Masked arrays

---

## 19. How does Pandas simplify time series analysis?
It offers:
- Datetime indexing
- Resampling and frequency conversion
- Rolling statistics
- Easy time-based slicing

---

## 20. What is the role of a pivot table in Pandas?
Pivot tables summarize data by grouping and aggregating — useful for reports and dashboards.

---

## 21. Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy slicing returns a **view** (no copy), thanks to contiguous memory layout, while Python list slicing creates a new object.

---

## 22. What are some common use cases for Seaborn?
- Distribution analysis (histograms, KDE)
- Correlation heatmaps
- Categorical data (bar, box, violin plots)
- Regression plots
- EDA

---


```python
#  1. How do you create a 2D NumPy array and calculate the sum of each row

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("2D Array:\n", arr)
print("Sum of each row:", row_sum)
```

```python
#  2. Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Marks': [85, 90, 78]
})
mean_marks = df['Marks'].mean()
print("Mean of Marks column:", mean_marks)
```

```python
# 3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns

data = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#  5. Generate a bar plot using Plotly
import plotly.express as px

bar_data = pd.DataFrame({
    'Fruits': ['Apples', 'Bananas', 'Cherries'],
    'Quantity': [10, 15, 7]
})
fig = px.bar(bar_data, x='Fruits', y='Quantity', title="Fruit Quantity Bar Plot")
fig.show()
```

```python
#  6. Create a DataFrame and add a new column based on existing column
df = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Price': [100, 150, 200]
})
df['Discounted Price'] = df['Price'] * 0.9
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print("Element-wise multiplication:", result)
```

```python
#  8. Create a line plot with multiple lines using Matplotlib
x = [0, 1, 2, 3]
y1 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2', linestyle='--')
plt.title("Multiple Lines Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid()
plt.show()
```
